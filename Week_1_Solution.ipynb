```python
#Excercise 2 Solution
# Import torch
import torch

# Create random tensor
A = torch.rand(size=(7, 7))
A, A.shape
```

```python
#Excercise 3 Solution

# Create another random tensor
B = torch.rand(size=(1, 7))
# C = torch.matmul(A, B) # will error because of shape issues
C = torch.matmul(A, B.T) # no error because of transpose
C, C.shape
```

```python
#Excercise 4 Solution

# Set manual seed
torch.manual_seed(0)

# Create two random tensors
A = torch.rand(size=(7, 7))
B = torch.rand(size=(1, 7))

# Matrix multiply tensors
C = torch.matmul(A, B.T)
C, C.shape
```

```python
# Excercise 5 Solution

# Set random seed on the GPU
torch.cuda.manual_seed(1234)
```

```python
# Excercise 6 Solution

# Set random seed
torch.manual_seed(1234)

# Check for access to GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")

# Create two random tensors on GPU
tensor_A = torch.rand(size=(2,3)).to(device)
tensor_B = torch.rand(size=(2,3)).to(device)
tensor_A, tensor_B
```

```python
# Excercise 7 Solution

# Perform matmul on tensor_A and tensor_B

# tensor_C = torch.matmul(tensor_A, tensor_B) # won't work because of shape error

tensor_C = torch.matmul(tensor_A, tensor_B.T)
tensor_C, tensor_C.shape
```

```python
# Excercise 8 Solution

# Find max
max = torch.max(tensor_C)

# Find min
min = torch.min(tensor_C)
max, min
```

```python
# Excercise 9 Solution

# Find arg max
arg_max = torch.argmax(tensor_C)

# Find arg min
arg_min = torch.argmin(tensor_C)
arg_max, arg_min
```
